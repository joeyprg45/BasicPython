## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
csv_file_path = 'C:/Users/jok/Desktop/PG/BasicPython/data/hist_data.csv'


df = pd.read_csv(csv_file_path)
a_data = df["aデータ"]
b_data = df["bデータ"]


plt.hist(a_data, bins=50, alpha=0.5, label="aデータ", color='orange')
plt.hist(b_data, bins=50, alpha=0.5, label="bデータ", color="lightblue")

plt.title("aデータとbデータの温度に関する度数分布")
plt.ylabel("度数")
plt.xlabel("温度(℃)")

plt.ylim(0,28)
plt.xlim(0,52)

plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
csv_file_path = 'C:/Users/jok/Desktop/PG/BasicPython/data/scatter_data.csv'

df = pd.read_csv(csv_file_path)

class0 = df[df["class"] == 0]
class1 = df[df["class"] == 1]
class2 = df[df["class"] == 2]

plt.figure(figsize=(4, 2.5))

plt.scatter(class0["length (cm)"], class0["width (cm)"], label="class0", color='red' ,s=20)
plt.scatter(class1["length (cm)"], class1["width (cm)"], label="class1", color="blue" ,s=20)
plt.scatter(class2["length (cm)"], class2["width (cm)"], label="class1", color="green" ,s=20)

plt.title("クラス0~2のlengthとwidthに関する散布図")
plt.ylabel("width(cm)")
plt.xlabel("length(cm)")

plt.ylim(1.9, 4.53)
plt.xlim(4.1, 8.1)

plt.legend(fontsize=7)

plt.grid(True)

plt.xticks([4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8])

plt.show()

```

使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
csv_file_path = 'C:/Users/jok/Desktop/PG/BasicPython/data/line_data.csv'

df = pd.read_csv(csv_file_path)

plt.figure(figsize=(4,2.5))

period = df["期間"]
cost = df["費用"]
count= df["件数"]
print(df)

#-----------------------------------------------
plt.plot(period, count , color="blue")
plt.scatter(period, count , color="blue")

plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')

plt.xticks(rotation=30, fontsize='small')       
plt.yticks([105,110,115,120,125,130,135,140])   

plt.grid(True)
plt.show()

#-----------------------------------------------------------

plt.bar(period, cost, color='blue')
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用(円)')

plt.xticks(rotation=30, fontsize='small') 

plt.grid(True)
plt.show()






```
